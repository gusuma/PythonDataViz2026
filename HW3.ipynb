## HW คำนวณเวลาเป็นวินาทีของเวลาต่อไปนี้โดยใช้คำสั่ง Split()ช่วย (print ออกมาให้สวยงาม)

1. 12:30:15  
2. 13:41:07

1.   12:53:15
2.   00:59:25

1.  11:11:11
2.   16:06:09









```python
times = [
    '12:30:15',
    '13:41:07',
    '12:53:15',
    '00:59:25',
    '11:11:11',
    '16:06:09'
]

for t_str in times:
    parts = t_str.split(':')

    hours = int(parts[0])
    minutes = int(parts[1])
    seconds = int(parts[2])

    total_seconds = hours * 3600 + minutes * 60 + seconds

    print(f"เวลา {t_str} มีค่าเท่ากับ {total_seconds:,} วินาที")
```

##ให้สร้างฟังก์ชั่นคำนวณเวลาเป็นวินาทีของเวลาใดๆ และ print ออกมาให้สวยงาม

```python
def calculate_total_seconds(time_str):
    """Calculates the total number of seconds from a time string (HH:MM:SS)."""
    parts = time_str.split(':')
    hours = int(parts[0])
    minutes = int(parts[1])
    seconds = int(parts[2])
    total_seconds = hours * 3600 + minutes * 60 + seconds
    print(f"เวลา {time_str} มีค่าเท่ากับ {total_seconds:,} วินาที")

```

##ให้ สร้าง list ของ เวลา ['12:30:15','13:41:07',....] แล้ววนลูปเรียกฟังก์ชั่นคำนวณเวลาเป็น วินาที

```python
times_list = [
    '12:30:15',
    '13:41:07',
    '12:53:15',
    '00:59:25',
    '11:11:11',
    '16:06:09'
]

for time_item in times_list:
    calculate_total_seconds(time_item)
```
